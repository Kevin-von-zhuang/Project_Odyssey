```python
python -m venv venv_name
```

```python
venv\Scripts\activate.bat
```

## *git*

用vscode打开新建项目文件夹的最外层，同时在github里新建一个库，将新建库时提供的地址告诉ai,直接使用ai完成与github的联结。

如果一个项目里有两个文件夹需要分别联结不同的github库，则应当将这两个库置于同一水平结构下。即无关独立仓库必须同级并列，左侧源代码管理（SCM）会分开列出两个仓库，各自有独立变更列表、分支、提交、push/pull 按钮。外层只是存放目录，本身不 git init，内部各个文件夹各自是独立仓库、各绑一个 GitHub，此时用同级并列多仓库结构。
