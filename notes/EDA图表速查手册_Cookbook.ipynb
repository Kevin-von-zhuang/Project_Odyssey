# 数据分析 EDA 图表速查手册（Cookbook）

## 一、单变量图表（仅分析单独一列，无其他字段参与）

### 1. 直方图 Histogram

#### 适配变量
连续数值变量（销售额、利润、运费、订单数量）

> 示例：全部订单销售额分布直方图

#### 核心价值
1. 将数值分段，直观展示每个区间样本绝对数量
2. 快速定位数据集中区间、识别数值断层、空白区间
3. 粗略观察分布形态：单峰/双峰、左偏/右偏

#### 局限性
1. 分箱宽度 `bins` 会大幅改变视觉效果，易造成误判
2. 无法平滑展示整体分布轮廓
3. 难以精准识别少量极端异常值

#### EDA 环节
3.1 连续数值变量单变量分布探查，与 KDE、箱线图配套使用

### 2. KDE 核密度曲线图

#### 适配变量
连续数值变量，常叠加在直方图上联合展示

> 示例：月度利润平滑密度曲线

#### 核心价值
1. 消除直方图分箱带来的视觉偏差，输出平滑连续分布
2. 清晰区分单峰、双峰、多峰分层数据
3. 通过左右长尾直观判断左偏/右偏，搭配均值、中位数线判断是否近似正态
4. 提前判断是否需要对数、标准化特征转换

#### 局限性
1. 纵轴为密度，无法直接读取真实样本数量
2. 极端异常长尾会压缩主体分布，看不清核心区间

#### EDA 环节
3.1 连续变量分布形态判定、偏态识别

### 3. 箱线图 Boxplot

#### 适配变量
连续数值变量，支持多组并列对比

> 示例：不同产品大类利润分布箱线图

#### 核心价值
1. 可视化四分位数、中位数、正常数据上下边界
2. 散点直接标记所有异常离群值，精准定位极端数据
3. 抗异常值干扰，多组并列时快速对比各组中心、离散程度

#### 局限性
1. 完全无法判断分布是正态、双峰还是单峰
2. 不能体现每个区间样本多少

#### EDA 环节
3.1 连续变量异常值筛查；多变量分组对比辅助工具

### 4. 分类频次柱状图

#### 适配变量
无序分类、有序等级、布尔二值变量

> 示例：各客户细分市场订单频次柱状图

#### 核心价值
1. 柱子高度代表样本绝对数量，直观对比各类体量差距
2. 可标注百分比，快速识别头部主流、尾部小众类别
3. 支持高基数多类别完整展示

#### 局限性
1. 不直观表达整体占比结构，份额需手动计算
2. 类别过多时横轴标签拥挤，可读性下降

#### EDA 环节
3.2 分类离散变量单变量频次统计分析

### 5. 饼图 / 环形图

#### 适配变量
低基数无序分类（类别 ≤ 6 个效果最佳）

> 示例：三大产品大类订单占比环形图

#### 核心价值
1. 直观展示各板块占整体 100% 的份额结构，快速识别核心业务板块

#### 局限性
1. 类别超过 6 个会标签拥挤、难以精确对比大小
2. 只能看占比，无法展示样本绝对数量
3. 不适合有序等级变量

#### EDA 环节
3.2 分类变量业务结构占比展示，用于业务汇报

## 二、双变量图表（两个字段交叉关联分析）

### 1. 散点图 Scatter Plot

#### 适配变量
两个连续数值变量

> 示例：销售额 vs 利润散点图

#### 核心价值
1. 判断两变量线性关系：正相关 / 负相关 / 无相关
2. 识别非线性趋势、聚类分组、局部异常点
3. 可增加颜色区分第三分类变量，拓展三维信息

#### 局限性
1. 数据量极大时会出现点重叠遮挡，看不清密度
2. 无法量化相关强弱

#### EDA 环节
4.1 数值变量相关性初步探查

### 2. 分组箱线图

#### 适配变量
X：分类变量；Y：连续数值变量

> 示例：不同运输方式的订单销售额分组箱线图

#### 核心价值
1. 横向对比不同类别下数值分布、中位数、异常值差异
2. 快速判断某类业务是否普遍盈利/亏损

#### 局限性
1. 无法看出每组内部样本数量多少

#### EDA 环节
4.2 分类 + 数值交叉分组对比

### 3. 分组柱状图 / 堆叠柱状图

#### 适配变量
X：分类变量；Y：连续数值指标

> 示例：各年度销售额、利润分组柱状图；各年份产品大类堆叠销售额图

#### 核心价值
- **分组柱状**：同一类别下多指标数值横向对比
- **堆叠柱状**：展示每个大类内部细分结构占比变化

#### 局限性
1. 堆叠柱状难以精准对比不同组别内部细分数值
2. 类别过多会视觉杂乱

#### EDA 环节
4.2 分类与指标交叉对比、结构拆解

### 4. 热力图 Heatmap

#### 适配变量
1. 全部连续数值相关系数矩阵
2. 分类 X × 分类 Y 交叉频次/指标汇总

> 示例：品类 & 市场订单量热力图、特征相关性热力图

#### 核心价值
1. 颜色深浅量化数值大小，快速定位高/低数值组合
2. 一眼识别强相关、弱相关特征
3. 快速挖掘交叉维度优质/薄弱业务板块

#### 局限性
1. 仅展示静态数值，无法体现变化趋势
2. 色阶主观，容易误判数值差距

#### EDA 环节
4.3 特征相关性分析、双分类交叉透视分析

## 三、时序专用图表（时间变量 + 数值指标，特殊双变量）

### 1. 折线图 / 面积图

#### 适配变量
X：时间（年/月/日）；Y：连续业务指标

> 示例：月度销售额折线面积图

#### 核心价值
1. 清晰展示长期增长/下滑趋势
2. 识别周期性淡旺季、月度波动高峰低谷
3. 面积图强化整体体量感知

#### 局限性
1. 多指标叠加过多会线条杂乱
2. 无法直观对比同期各类别占比

#### EDA 环节
3.3 时间维度单变量趋势分析、时序规律挖掘

### 2. 双轴时序图 twinx

#### 适配变量
X：时间；两组量级差距极大的连续指标

> 示例：年度订单量（左轴）+ 平均折扣率（右轴）

#### 核心价值
1. 同一画布联动观察两个量级差异巨大的指标
2. 直观判断指标联动关系（折扣拉高是否带动订单增长）

#### 局限性
1. 双坐标轴容易误导视觉对比

#### EDA 环节
3.3 多指标时序联动分析

## 四、多变量综合图表（3 个及以上变量同时可视化）

### 1. 平行坐标图

#### 适配变量
多条连续数值特征

> 示例：订单销售额、利润、折扣、订单数量多维度平行图

#### 核心价值
1. 同时展示多条数值特征分布
2. 识别样本聚类、特征区间分布规律

#### 局限性
1. 维度过多线条密集，可读性差

#### EDA 环节
高维数据整体分布探查

### 2. Q-Q 正态分位图

#### 适配变量
单条连续数值（统计校验专用）

> 示例：利润 Q-Q 图检验正态性

#### 核心价值
1. 判断数据是否近似正态分布
2. 偏离对角线说明存在强偏态，需要做对数变换

#### 局限性
仅用于正态校验，无业务直观解读

#### EDA 环节
连续变量分布正态性检验，特征工程前置判断

### 3. 残差图

#### 适配变量
回归模型预测值、真实残差（建模后分析）

> 示例：销售额回归模型残差分布图

#### 核心价值
1. 判断残差是否随机均匀分布
2. 识别模型非线性偏差、异常干扰样本

#### 局限性
仅建模后使用，不属于基础 EDA

#### EDA 环节
建模后模型效果诊断
